```python
class Patient:
    def __init__(self, name, age, gender, blood_pressure, temperature):
        self.name = name
        self.age = age
        self.gender = gender
        self.blood_pressure = blood_pressure
        self.temperature = temperature

    def display_patient_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        print(f"Blood pressure: {self.blood_pressure}")
        print(f"Temperature: {self.temperature}")

    def is_hypertensive(self):
        systolic,diastolic = self.blood_pressure
        if systolic >= 140 or diastolic >= 90:
            return True
        else:
            return False 

    def is_feverish(self):
        temp = self.temperature
        if temp >=37.5:
            return True
        else:
            return False 
            
            
patient1_data = Patient("John Doe", 45, "Male", (140, 85), 36.8)
patient2_data = Patient("Jane Smith", 32, "Female", (130, 88), 37.6)

print ("Patient 1 information: ")
patient1_data.display_patient_info()
print()

print ("Patient 2 information: ")
patient2_data.display_patient_info()
print()

print (f"Patient 1 is Hypertensive:{patient1_data.is_hypertensive()}")
print (f"Patient 1 is Feverish: {patient1_data.is_feverish()}")
print()

print (f"Patient 2 is Hypertensive: {patient2_data.is_hypertensive()}")
print (f"Patient 2 is Feverish: {patient2_data.is_feverish()}")
print()


        
    
```
